```python
def along_track_distance(latitude,longtitude,reference):
# Calculate along track distance
# Parameters:
#       lat and lon for each photon 
#       (lat,lon) array for a referenec point
# Returns
    # Calculate distance for each successive point
    distances = np.zeros(len(latitude))
    for i in range(0,len(latitude)):
        distances[i] = geodesic(reference, [latitude[i], longtitude[i]]).meters

    return distances
```

```python
def closest_indices(longitudes_1, longitudes_2):
    """
    Find the index of the closest longitude in longitudes_1 to each longitude in longitudes_2.

    Parameters:
        longitudes_1 (list or numpy array): List of longitudes to compare against.
        longitudes_2 (list or numpy array): List of longitudes for which to find the closest indices.

    Returns:
        list: List of indices in longitudes_1 closest to each longitude in longitudes_2.
    """
    indices = []
    for lon_2 in longitudes_2:
        # Calculate the absolute differences between lon_2 and all values in longitudes_1
        abs_diff = np.abs(np.array(longitudes_1) - lon_2)
        # Find the index of the minimum absolute difference
        closest_index = np.argmin(abs_diff)
        indices.append(closest_index)
    return indices

def find_index_lead(distance, center_index, tolerance=10):
    """
    Find the index intervals with given indexes as centers and a specified longitude tolerance.

    Parameters:
        longitudes (list or numpy array): List of longitudes.
        center_indexes (list or numpy array): List of indexes to use as the centers of the intervals.
        tolerance (float): Meter tolerance to determine the interval boundaries. Default is 10.

    Returns:
        list: A list containing tuples of start and end indices for each center index.
    """
    intervals = []
    for center_idx in center_index:
        center_lead = distance[center_idx]
        start_index = center_idx
        end_index = center_idx

        # Find the start index of the interval
        while start_index > 0 and center_lead - distance[start_index] <= tolerance:
            start_index -= 1

        # Find the end index of the interval
        while end_index < len(distance) - 1 and distance[end_index] - center_lead <= tolerance:
            end_index += 1

        intervals.append((start_index, end_index))

    return intervals
```

```python
def plot_lead_histogram(height_data, intervals):
    """
    Plot a histogram of height data within the specified index intervals.

    Parameters:
        height_data (list or numpy array): List of height data corresponding to longitude values.
        intervals (list): List of tuples containing start and end indices for each interval.

    Returns:
        None
    """
    # Flatten the intervals to get all the indices
    all_indices = [idx for interval in intervals for idx in range(interval[0], interval[1] + 1)]

    # Remove any duplicate indices
    unique_indices = list(set(all_indices))

    # Extract the corresponding height values
    heights_within_intervals = [height_data[idx] for idx in unique_indices]

    # Plot histogram
    plt.hist(heights_within_intervals, bins=1000, density=True, alpha=0.5, color='skyblue', edgecolor='black')
    plt.title('Lead Height Data Histogram')
    plt.xlabel('Height')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
    
    group_characteristics = {}
    group_characteristics['std'] = np.std(heights_within_intervals)
    group_characteristics['mean'] = np.mean(heights_within_intervals)
    group_characteristics['median'] = np.median(heights_within_intervals)
    group_characteristics['peak'] = heights_within_intervals[np.argmax(heights_within_intervals)]

    print("Characteristics of the group:")
    for key, value in group_characteristics.items():
        print(f"{key}: {value}")
```

```python
def calculate_distribution_characteristics(histograms):
    """
    Calculate characteristics of the distribution for each histogram.

    Parameters:
        histograms (list of numpy arrays): List of histograms.

    Returns:
        list: A list containing dictionaries with characteristics for each histogram.
    """
    characteristics = []

    for hist in histograms:
        # Calculate mean
        mean = np.mean(hist)
        
        # Calculate standard deviation
        std_dev = np.std(hist)
        
        # Calculate skewness
        skewness = np.mean((hist - mean) ** 3) / (std_dev ** 3)
        
        # Calculate kurtosis
        kurtosis = np.mean((hist - mean) ** 4) / (std_dev ** 4) - 3

        #Calculate peak
        peak = np.mean(argmax(hist))
        characteristics.append({
            'mean': mean,
            'standard_deviation': std_dev,
            'skewness': skewness,
            'kurtosis': kurtosis,
            'peak':peak
        })

    return characteristics
def calculate_group_characteristics(histograms):
    """
    Calculate characteristics of a group of histograms.

    Parameters:
        histograms (list of numpy arrays): List of histograms.

    Returns:
        dict: A dictionary containing the mean characteristics of the group.
    """
    # Calculate characteristics for each histogram
    characteristics_list = calculate_distribution_characteristics(histograms)
    
    # Calculate mean characteristics
    mean_characteristics = {
        'mean_std': np.mean([c['std'] for c in characteristics_list]),
        'mean_width': np.mean([c['width'] for c in characteristics_list]),
        'mean_skewness': np.mean([c['skewness'] for c in characteristics_list]),
        'mean_kurtosis': np.mean([c['kurtosis'] for c in characteristics_list]),
        'mean_peak': np.mean([c['peak'] for c in characteristics_list])
    }
    
    return mean_characteristics

def classify_histogram_as_lead(histogram, group_characteristics, tolerance=0.1):
    """
    Classify a histogram as a lead based on group characteristics and tolerance.

    Parameters:
        histogram (numpy array): Histogram to classify.
        group_characteristics (dict): Mean characteristics of the group.
        tolerance (float): Tolerance level for matching characteristics. Default is 0.1.

    Returns:
        bool: True if the histogram is classified as a lead, False otherwise.
    """
    # Calculate characteristics of the histogram
    histogram_characteristics = calculate_distribution_characteristics([histogram])[0]

    # Check if the histogram characteristics are within the tolerance of the group characteristics
    for characteristic, value in histogram_characteristics.items():
        if abs(value - group_characteristics[f'mean_{characteristic}']) > tolerance:
            return False

    return True
```

```python
def calculate_average_peak(height_data, center_indexes, distance, tolerance=10):
    """
    Calculate the average peak height of all the center leads within a 10km segment.

    Parameters:
        track_data (list or numpy array): Height measurements along the track.
        center_indexes (list): List of indexes corresponding to the center of leads.
        distance (list or numpy array): List of distances.
        tolerance (float): Distance tolerance to determine the interval boundaries. Default is 10.

    Returns:
        float or None: The average peak height of the center leads within the segment, or None if no suitable leads are found.
    """
    peak_heights = []

    # Find the index intervals for each center lead within the segment
    intervals = find_index_lead(distance, center_indexes, tolerance)

    # Iterate over the intervals and find the peak height for each lead
    for start_idx, end_idx in intervals:
        segment_data = height_data[start_idx:end_idx + 1]  # Include the end index
        hist, bin_edges = np.histogram(segment_data, bins=100, density=True)
        peaks, _ = find_peaks(hist, prominence=0.1)

        # Sort the peaks by their heights
        sorted_peaks = sorted(peaks, reverse=True)

        # Take the highest peak to find the mode
        if sorted_peaks:
            mode_index = sorted_peaks[0]
            mode_height = bin_edges[mode_index]
            peak_heights.append(mode_height)

    # Calculate the average peak height
    if peak_heights:
        print(peak_heights)
        average_peak_height = np.mean(peak_heights)
        return average_peak_height
    else:
        return None
```

```python
def calculate_sea_surface(track_data, center_indexes, segment_length=10):
    """
    Calculate the sea surface from a 10km track of height measurements.

    Parameters:
        track_data (list or numpy array): Height measurements along the track.
        center_indexes (list): List of indexes corresponding to the center of leads.
        segment_length (int): Length of each segment in kilometers. Default is 10km.

    Returns:
        float or None: The estimated sea surface height, or None if no suitable leads are found.
    """
    sea_surface_height = []

    # Find the index of the first lead
    first_lead_idx = center_indexes[0]

    # Calculate the start and end index of the track segment
    start_idx = max(0, first_lead_idx - segment_length * 500)  # 10km track segment (500 measurements)
    end_idx = min(len(track_data), first_lead_idx + segment_length * 501)  

    # Extract the track segment
    segment_data = track_data[start_idx:end_idx]

    # Plot histogram
    hist, bin_edges = np.histogram(segment_data, bins=100, density=True)

    # Find the highest peak of the density histogram
    peak_idx = sorted(find_peaks(hist,prominence=0.1),reverse=True)[0]
    sea_surface_height = bin_edges[peak_idx]

    return sea_surface_height

def calculate_freeboard(track_data, sea_surface):
    """
    Calculate sea ice freeboard from height measurements and sea surface.

    Parameters:
        track_data (list or numpy array): Height measurements along the track.
        sea_surface (float): Estimated sea surface height.

    Returns:
        numpy array: Sea ice freeboard values.
    """
    return np.array(track_data) - sea_surface

# Example usage
track_data = np.random.randn(1000)  # Example height measurements
center_indexes = [100, 300, 500]  # Example list of center indexes corresponding to leads
sea_surface = calculate_sea_surface(track_data, center_indexes)
if sea_surface is not None:
    freeboard = calculate_freeboard(track_data, sea_surface)
    print("Sea Surface Height:", sea_surface)
    print("Sea Ice Freeboard:", freeboard)
else:
    print("No suitable leads found.")
```

```python
def find_lead_indexes_within_interval(center_indexes, alongtrack_distance, interval_start_idx, interval_end_idx, lead_range=7.5):
    """
    Find the indexes of leads within a specified range around the given center indexes within a distance interval.

    Parameters:
        center_indexes (array_like): Array of lead center indexes.
        alongtrack_distance (array_like): Array of alongtrack distances.
        interval_start_idx (int): Start index of the distance interval.
        interval_end_idx (int): End index of the distance interval.
        lead_range (float): Range around the center index to search for leads. Default is 7.5 meters.

    Returns:
        list: A list containing arrays of indexes corresponding to leads within the specified range for each center index.
    """
    lead_indexes_list = []
    for center_idx in center_indexes:
        if interval_start_idx <= center_idx <= interval_end_idx:
            # Get the start and end distances for the buffer around the center index
            lead_start_dist = alongtrack_distance[center_idx] - lead_range
            lead_end_dist = alongtrack_distance[center_idx] + lead_range
            
            # Find the start and end indexes within the interval
            lead_start_idx = max(interval_start_idx, np.searchsorted(alongtrack_distance, lead_start_dist, side='left'))
            lead_end_idx = min(interval_end_idx, np.searchsorted(alongtrack_distance, lead_end_dist, side='right'))
            
            # Add the indexes within the specified range to the list
            lead_indexes_list.append(np.arange(lead_start_idx, lead_end_idx + 1))
    
    return lead_indexes_list
```

```python
def find_lead_histograms_and_char(along_track_distance,height_data, lead_centers, interval_start,interval_end):
    histograms = []
    hist_bins=[]
    avg_peak_density = []
    all_peak_heights = []
    avg_peak_height = []
    lead_peaks = []
    all_peak_density = []
    all_peak_density_std = []

    for center_idx in lead_centers:
        if interval_start <= along_track_distance[center_idx] <= interval_end:
        # Extract 150 photons around the lead center
            start_idx = max(center_idx - 75, 0)
            end_idx = min(center_idx + 75, len(along_track_distance))
            lead_heights = height_data[start_idx:end_idx]

        # Compute the histogram of the segment
            hist, bin_edges = np.histogram(lead_heights, bins=100, density=True)
            histograms.append(hist)
            hist_bins.append(bin_edges)
        # Find peaks in the histogram
            peaks, _ = find_peaks(hist, prominence=0.3)
            lead_peaks.append(peaks)
            peak_heights = bin_edges[peaks]
        # Save peak values
            peak_density = hist[peaks]
            all_peak_density.extend(peak_density)
            all_peak_density_std.append(np.std(peak_density))
            dist_mean = np.mean(bin_edges)
            # Calculate average height value of peaks
            peak_heights = bin_edges[peaks]

            # Save peak values
            all_peak_heights.append(peak_heights)

    # Calculate overall averages
    avg_peak_density = np.mean(all_peak_density)
    avg_peak_density_std = np.mean(all_peak_density_std)
    avg_peak_height = np.mean(all_peak_heights)
    avg_dist_mean = np.mean(dist_mean)
    return histograms, hist_bins, lead_peaks, avg_peak_density, avg_peak_density_std, avg_peak_height, avg_dist_mean


# Assuming alongtrack_distance and height are numpy arrays containing distance and height data
# center_indexes is a list of lead center indexes
# interval_start and interval_end define the alongtrack distance interval
histograms, hist_bins, lead_peaks,avg_lead_density,std_lead_density,avg_lead_h, avg_dist_mean = find_lead_histograms_and_char(distances, filtered_ph_h, unique_indices_lead_center, interval_start=95000, interval_end=105000)
#Intervals [50000:60000], [95000,105000]

print(avg_lead_density,avg_dist_mean, std_lead_density,avg_lead_h)

# Plot the histograms and lead peaks
for hist,bins,peaks in zip(histograms, hist_bins, lead_peaks):
    plt.plot(bins[:-1], hist)  # Use bin edges as x-values
    plt.plot(bins[peaks], hist[peaks], 'ro')  # Mark peaks on the histogram
plt.xlabel('Height')
plt.ylabel('Density')
plt.title('Histograms of Individual Leads')
plt.show()
```

```python
def find_surface_and_subsurface_modes(distance, height, segment_length=20, prominence_threshold=0.2, bins=100, mode_distance_threshold=0):
    """
    Find the surface and subsurface modes within each 10-meter segment of (distance, height) data
    using peak detection on the histogram of height data.
    
    Parameters:
        distance (array_like): Array of distance measurements.
        height (array_like): Array of corresponding height measurements.
        segment_length (float): Length of each segment in meters. Default is 20 meters.
        prominence_threshold (float): Minimum prominence of peaks in the histogram. Default is 0.2
        bins (int): Number of bins for histogram computation. Default is 100.
        mode_distance_threshold (int): Maximum distance between modes to consider them as the same mode. Default is 0 meters.
    Returns:
        tuple: Tuple containing arrays of surface and subsurface modes and their corresponding segment indices.
    """
    surface_modes = []
    subsurface_modes = []
    startoflake_points = []
    endoflake_points = []
    segment_indices = []
    #Make distance start at 0
    distance = distance-distance[0]
    prev_subsurface_mode = None
    prev_surface_mode = None
    start_idx = 0
    while start_idx < len(distance):
        end_idx = start_idx
        while end_idx < len(distance) and distance[end_idx] - distance[start_idx] <= segment_length:
            end_idx += 1

        segment_distance = distance[start_idx:end_idx]
        segment_heights = height[start_idx:end_idx]

        if len(segment_heights) > 0:
            # Compute histogram of segment_heights
            hist, bin_edges = np.histogram(segment_heights, bins=bins,density=True)
            # Define kernel for convolution
            # kernel = np.array([0.25, 1,0.25])  # Kernel weights for neighboring bins

            # Convolve histogram with kernel
            # smoothed_hist = np.convolve(hist, kernel, mode='same')  # 'same' mode to keep the same size

            # Find peaks in the histogram
            peaks, _ = find_peaks(hist, prominence=prominence_threshold,distance=3,plateau_size=1) 

            # Select the surface mode and potentially the subsurface mode
            if len(peaks) > 0:
                # Get heights of peaks
                peak_heights = hist[peaks]

                # Sort peaks by their height values
                sorted_peaks = sorted(peaks, key=lambda idx: bin_edges[idx], reverse=True)

                # Consider only the two highest peaks if they exist
                if len(sorted_peaks) > 1:
                    surface_mode_idx = sorted_peaks[0]
                    surface_mode = bin_edges[surface_mode_idx]

                    subsurface_mode_idx = sorted_peaks[1]
                    subsurface_mode = bin_edges[subsurface_mode_idx]

                    surface_modes.append(surface_mode)
                    subsurface_modes.append(subsurface_mode)

                    # Save startoflake point if previous segment did not have both modes
                    if prev_subsurface_mode is None:
                        startoflake_points.append(segment_distance[0])

                elif len(sorted_peaks) == 1:
                    surface_mode_idx = sorted_peaks[0]
                    surface_mode = bin_edges[surface_mode_idx]

                    #Save end of lake point if previous segment had both modes 
                    if prev_subsurface_mode is not None: 
                        if surface_mode < prev_subsurface_mode:
                            subsurface_modes.append(surface_mode)
                            surface_modes.append(prev_surface_mode)
                            
                        if abs(surface_mode - prev_subsurface_mode) <= mode_distance_threshold:
                            subsurface_modes.append(surface_mode) 
                            surface_modes.append(prev_surface_mode)
                        else:
                            endoflake_points.append(segment_distance[0])
                            surface_modes.append(surface_mode)
                            subsurface_modes.append(None)
                    
                    elif prev_subsurface_mode is None:
                        surface_modes.append(surface_mode)
                        subsurface_modes.append(None)                                 
                else:
                    surface_modes.append(None)
                    subsurface_modes.append(None)

            else:
                surface_modes.append(prev_surface_mode)
                subsurface_modes.append(prev_subsurface_mode)
            
            segment_indices.append((start_idx, end_idx))

        else:
            surface_modes.append(None)
            subsurface_modes.append(None)

            segment_indices.append((start_idx, end_idx))
        #Update prev_surface_mode
        if len(surface_modes) > 0:
            prev_surface_mode = surface_modes[-1]
        else:
            prev_surface_mode = None
        # Update prev_subsurface_mode
        if len(subsurface_modes) > 0:
            prev_subsurface_mode = subsurface_modes[-1]
        else:
            prev_subsurface_mode = None

        start_idx = end_idx
        
        # Calculate the meltpond diameter and depth

        # Corrected depth of subsurface modes is given by the refractive index of water: cor_sub_modes = sub_modes * 0.75 #see Ellen M Buckley

    return np.array(surface_modes), np.array(subsurface_modes), np.array(startoflake_points), np.array(endoflake_points), segment_indices

def calculate_lake_length_and_depth(startoflake_points, endoflake_points, subsurface_modes, surface_modes, distances):
    lake_lengths = []
    lake_depths = []

    for start, end in zip(startoflake_points, endoflake_points):
        if start is None or end is None:
            continue

        # Calculate lake length
        lake_length = end - start
        lake_lengths.append(lake_length)

    for subsurface_mode, surface_mode in zip(subsurface_modes, surface_modes):
        if subsurface_mode is None or surface_mode is None:
            continue

        # Calculate lake depth
        lake_depth = (surface_mode - subsurface_mode) * 1.00029 / 1.33567
        lake_depths.append(lake_depth)

    # Calculate average length and depth
    coverage = np.sum(lake_lengths) / distances[-1] *100
    avg_length = np.sum(lake_lengths) / len(lake_lengths) if lake_lengths else 0
    avg_depth = np.sum(lake_depths) / len(lake_depths) if lake_depths else 0

    print("Total lake coverage:",coverage,'%')
    print("Average Lake Length:", avg_length,'m')
    print("Average Lake Depth:", avg_depth*100,'cm')

    return lake_lengths, lake_depths, avg_length, avg_depth, coverage
```

```python
def find_surface_modes(distance, height, segment_length=20, bins=100):
    """
    Find the surface modes within each segment of (distance, height) data.

    Parameters:
        distance (array_like): Array of distance measurements.
        height (array_like): Array of corresponding height measurements.
        segment_length (float): Length of each segment in meters. Default is 20 meters.
        bins (int): Number of bins for histogram computation. Default is 100.

    Returns:
        numpy.ndarray: Array containing the surface modes for each segment.
    """
    surface_modes = []
    prev_surface_mode = None
    startoflake_points = []
    endoflake_points = []
    start_idx = 0
    while start_idx < len(distance):
        end_idx = start_idx + segment_length
        if end_idx > len(distance):
            end_idx = len(distance)

        segment_distance = distance[start_idx:end_idx]
        segment_heights = height[start_idx:end_idx]

        if len(segment_heights) > 0:
            # Compute histogram of segment_heights
            hist, bin_edges = np.histogram(segment_heights, bins=bins, density=True)
            # Find peaks in the histogram
            peaks, _ = find_peaks(hist, prominence=0.2)

            # Select the surface mode
            if len(peaks) == 0:
                # If no peaks found, use the previous segment's surface mode
                surface_modes.append(prev_surface_mode)
            elif len(peaks) == 1:
                # If only one peak found, segment into 5m subsegments and find peaks in each
                subsegment_length = 5
                num_subsegments = int(segment_length / subsegment_length)
                subsegment_modes = []
                for i in range(num_subsegments):
                    subsegment_heights = segment_heights[i * subsegment_length : (i + 1) * subsegment_length]
                    subsegment_hist, _ = np.histogram(subsegment_heights, bins=bins, density=True)
                    subsegment_peaks, _ = find_peaks(subsegment_hist, prominence=0.2)
                    if len(subsegment_peaks) > 0:
                        subsegment_mode = bin_edges[subsegment_peaks[0]]
                        subsegment_modes.append(subsegment_mode)
                # Take the 4 highest density peaks from subsegments
                subsegment_modes.sort(reverse=True)
                surface_modes.append(subsegment_modes[:4])
            else:
                # If multiple peaks found, use the highest peak as the surface mode
                surface_mode = bin_edges[peaks[0]]
                surface_modes.append(surface_mode)
                # Save startoflake point if previous segment did not have multiple peaks
                if prev_surface_mode is None:
                    startoflake_points.append(segment_distance[0])

        else:
            # If segment is empty, use the previous segment's surface mode
            surface_modes.append(prev_surface_mode)

        # Update prev_surface_mode
        prev_surface_mode = surface_modes[-1]

        start_idx = end_idx

    # Identify endoflake points
    for i in range(1, len(surface_modes)):
        if surface_modes[i] is not None and surface_modes[i - 1] is not None:
            endoflake_points.append(distance[i * segment_length])

    return np.array(surface_modes), np.array(startoflake_points), np.array(endoflake_points)

# Test the function
distance = np.arange(0, 200, 0.1)  # Example distance data
height = np.random.normal(loc=0, scale=1, size=len(distance))  # Example height data
segment_length = 20
surface_modes, startoflake_points, endoflake_points = find_surface_modes(distance, height, segment_length)
print(surface_modes)
print(startoflake_points)
print(endoflake_points)
```

```python
def interpolate_modes(surface_modes, subsurface_modes, segment_indices):
    interpolated_surface_modes = []
    interpolated_subsurface_modes = []

    for i in range(len(segment_indices)):
        surface_mode_value = surface_modes[i]
        start_idx, end_idx = segment_indices[i]
        segment_length = end_idx - start_idx

        if surface_mode_value is None:
            interpolated_surface_modes.extend([None] * segment_length)
        else:
            surface_mode_array = np.full(segment_length, surface_mode_value)
            interpolated_surface_modes.extend(surface_mode_array)

        subsurface_mode_value = subsurface_modes[i]

        if subsurface_mode_value is None:
            interpolated_subsurface_modes.extend([None] * segment_length)
        else:
            subsurface_mode_array = np.full(segment_length, subsurface_mode_value)
            interpolated_subsurface_modes.extend(subsurface_mode_array)

    return np.array(interpolated_surface_modes), np.array(interpolated_subsurface_modes)
```
